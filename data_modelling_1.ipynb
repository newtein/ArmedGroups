```python
col_drop = ["triadid", "UCDPdyadid", "tarnum_gw","nag_name"]
impute_median = ["yr_active", "foundyr", 'nagid_1',
       'nagid_2', 'nagid_3', 'nagid_4', 'nagid_5', 
       'nagleft', 'nagright', 'nagdem', 'nagauth', 'nagdict', 'nagmil',
       'nagtheo', 'nagobj_1', 'nagobj_2', 'nagobj_3', 'nagobj_4', 'nagobj_5',
       'nagobj_6', 'statesup', 's_safemem', 's_precision_sm', 's_safelead',
       's_precision_sl', 's_headquar', 's_precision_head', 's_traincamp',
       's_precision_tc', 's_training', 's_precision_train', 's_weaponlog',
       's_precision_wl', 's_finaid', 's_precision_finaid', 's_transport',
       's_precision_tp', 's_troop', 's_precision_tr', 's_other',
       's_precision_oth', 'defacto', 'ds_safemem', 'ds_precision_sm',
       'ds_safelead', 'ds_precision_sl', 'ds_headquar', 'ds_precision_head',
       'ds_traincamp', 'ds_precision_tc', 'ds_training', 'ds_precision_train',
       'ds_weaponlog', 'ds_precision_wl', 'ds_finaid', 'ds_precision_finaid',
       'ds_transport', 'ds_precision_tp', 'ds_other', 'ds_precision_oth',
       'domsup', 'domsup_p']
impute_mode = ["nageth", "nagrel"]
lambda_col = ["startdate1", "startdate2", "terr", "supporter"]
one_hot_col = ["target","terr", "partyname", "supporter"]
```

```python
df.drop(col_drop, axis=1, inplace=True)
```

```python
for col in impute_median+impute_mode:
    df[col] =df[col].fillna(df[col].mode()[0])
```

```python
def create_brd_year(brd):
    if not pd.isna(brd):
        brd = brd.split('/')[-1]
        if type(brd) in [int, float]:
            brd = '19' + brd if 21 <= int(brd) <= 99 else '20' + brd
        brd = brd.split('-')[0]
        return int(brd)
    else:
        return 0

df['startdate1'] =  df['startdate1'].apply(create_brd_year)
df['startdate2'] =  df['startdate2'].apply(create_brd_year)
```

```python
df['terr'] = df['terr'].fillna('No territory')
df['supporter'] = df['supporter'].fillna('No supporter')
```
